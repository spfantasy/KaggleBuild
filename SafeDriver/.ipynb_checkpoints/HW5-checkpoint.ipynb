### 2.AML Problem 1.7(a)

### toss coins with P(head) = P(error) = $\mu$, while obtaining k heads in N tosses, training error $\upsilon = \frac{k}{N}$

#### (a)Assume the sample size is 10, if $\mu$ = 0.05, compute the probabilty that at least one coin will have $\upsilon$ = 0 for case of 1/1,000/1,000,000 coins. Repeat for $\mu$ = 0.8

P($\upsilon$ = 0) = p(k = 0|N = 10, $\mu$ = 0.05) = $(0.95)^{10}$

P(at least one coin have $\upsilon$ = 0) = 1 - P(no coin have $\upsilon$ = 0)

P(at least one coin have $\upsilon$ = 0|1 coin, $\mu$ = 0.05) = 1 - $[1-(0.95)^{10}]^1$ = 0.5987

P(at least one coin have $\upsilon$ = 0|1,000 coin, $\mu$ = 0.05) = 1 - $[1-(0.95)^{10}]^{1000}$ $\approx$ 1

P(at least one coin have $\upsilon$ = 0|1,000,000 coin, $\mu$ = 0.05) = 1 - $[1-(0.95)^{10}]^{1000000}\approx$  1

P($\upsilon$ = 0) = p(k = 0|N = 10, $\mu$ = 0.8) = $(0.2)^{10}$

P(at least one coin have $\upsilon$ = 0|1 coin, $\mu$ = 0.8) = 1 - $[1-(0.2)^{10}]^1$ = 1e-7

P(at least one coin have $\upsilon$ = 0|1,000 coin, $\mu$ = 0.8) = 1 - $[1-(0.2)^{10}]^{1000}$ = 1e-4

P(at least one coin have $\upsilon$ = 0|1,000,000 coin, $\mu$ = 0.8) = 1 - $[1-(0.2)^{10}]^{1000000}$  = 0.097

#### (b)Take  the  scenario  of  part  (a),  for  the  case  of  1,000  coins ,and $\mu$ =0.05.   Consider  the  following interpretation  in applying  it  in a machine  learning setting.

##### (ii) What does the calculation in part (a), for 1000 coins and μ = 0.05, represent?

It means the probability if at least one hypothesis h in hypothesis space makes a 100% accuracy on the training set.

when that kind of situation happens, the chosen hypothesis g, will lead to v = $E_{in}(g)$ = 0

So it is also equals to P($E_{in}(g)$ = 0)

##### (iii) In this interpretation, take the most general version of the Hoeffding inequality in Ch. 1, State what $\mu$,v,M and N are.

$\mu$ means $E_{out}(g)$ for the given hypothesis. g is the certain hypothesis chosen.
It is the true error (true possibility of appearance of head)(out-of-sample error)

v means $E_{in}(g)$ for the given hypothesis. g is the certain hypothesis chosen.
It is the training error (percentage of appearance of head for a single coin)(in-sample error)

M is the number of hypothesis(number of coins)(number of bins), which is 1000

N is number of samples to evaluate a single hypothesis(number of tosses for a single coin), which is 10
